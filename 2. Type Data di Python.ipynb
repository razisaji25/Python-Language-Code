1. Lakukan 4 operasi aritmatika pada 2 variabel di bawah (tambah, kurang, kali, bagi), lalu simpan semua hasilnya didalam list dan tampilkan.

var_1 = 10

var_2 = 2

```python
#kerjakan soal no 1 pada sel ini
#keluaran yang diharapkan : [12,8,20,5]
Output = []
var_1 = 10
var_2 = 2
Output_1 = var_1 + var_2
Output_2 = var_1 - var_2
Output_3 = var_1 * var_2
Output_4 = var_1 / var_2

Output.append(Output_1)
Output.append(Output_2)
Output.append(Output_3)
Output.append(Output_4)

print(Output)
```
